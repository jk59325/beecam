```python
# takes videos from 'download to transcode' directory, chooses videos for today, appends them one after each other into a long video, 
# filename hive#-camera#-datetime.avi
import os
import time
from datetime import datetime
import getopt
import sys
from IPython.display import Image
from pathlib import Path
import os
import datetime
from send2trash import send2trash
import numpy as np
import cv2

curDir = 'F:\\downloads\\to transcode\\'
outDir = 'K:\\DCIM\\drone\\dcim\\100DSCIM\\bees\\'
handbrakePath = "C:\\Users\\jk\\imports\\HandBrakeCLI.exe"
included_extensions = ['avi','mp4']
width = 1920
height = 1080


today = time.strftime("%y%m%d")
#today = "200511"
print(today)

os.chdir(curDir)
fn = os.listdir(curDir)
dirlist = [fn for fn in fn
              if any(fn.endswith(ext) for ext in included_extensions)]
    
def todayVideo(dirlist):
    result = list()
    for f in dirlist:
        if today in f:
            result.append(f)
    return result


def transcode(inpath, outpath):
    #print(inpath)
    #print(outpath)
    cmd  = handbrakePath + ' -i "'
    cmd += inpath + '" -o "'
    cmd += outpath + '" '
    cmd += "--preset-import-file jonpreset.json"
    os.system(cmd)
    #print(cmd)
    #print(str(os.system(cmd)))

def rescale(frame):
    fheight = frame.shape[0]
    fwidth = frame.shape[1]
    sH = height / fheight  
    sW = width / fwidth
    if (sH < sW):
        fac = sH
    else:
        fac = sW
    dim = (int(fwidth*fac),int(fheight*fac))
    #print(frame.shape)
    #print(dim)
    return dim

def addBorder(frame,dim):
    value = [0,0,0]
    xH = height - dim[1]
    xW = width - dim[0]
    borderH = int(xH / 2)
    borderW = int(xW / 2)
    dst = cv2.copyMakeBorder(frame, borderH, borderH, borderW, borderW, cv2.BORDER_CONSTANT, None, value)
    return dst

def addCaption(frame,f):
    font                   = cv2.FONT_HERSHEY_SIMPLEX
    bottomLeftCornerOfText = (10,20)
    fontScale              = 0.7
    fontColor              = (255,255,255)
    lineType               = 2    
    
    cv2.putText(frame,f, 
        bottomLeftCornerOfText, 
        font, 
        fontScale,
        fontColor,
        lineType)
    return frame

global arduDict
arduDict = dict()
# ['hive2-ardu-200504', '0941', '132', '115', 'mp4']
def isArduRepeat(fpart):
    # check if ardu shot was done within the same hour and same day, return true
    # otherwise returns false
    global arduDict    
    #print(fpart)
    fpart = fpart.replace(".","_")
    fs = fpart.split("_")
    candidateHour = int(int(fs[1]) / 100)
    #print(fs)
    #()+1
    #print(arduDict)
    uniqueID = fs[0] + str(fs[-3]) + str(fs[-2])
    if uniqueID in arduDict:
        return True
        #if candidateHour == arduDict[uniqueID]:
        #    return True
    else:
        arduDict[uniqueID] = candidateHour
        return False
    #()+1
    return False

def limitTheArdu(dirlist):
    # only accept ardu shots in the same hour
    result = list()
    #print(dirlist)
    for f in dirlist:
        if "ardu" in f:
            if (isArduRepeat(f)):
                print('removed ' + f)
            else:
                #print('added ' + f)
                result.append(f)
        else:
            result.append(f)
    #print(result)
    return result
    

dirlist = todayVideo(dirlist)
print(dirlist)
dirlist = limitTheArdu(dirlist)
#dirlist = [f for f in dirlist if "ardu" not in f]
dirlist = [f for f in dirlist if "snapshot" not in f]
dirlist = [f for f in dirlist if "gem" not in f]
videofiles = sorted(dirlist)
print(videofiles)
print(len(videofiles))
#()+1
#i=len(dirlist)
#i = levenshtein(dirlist[0],dirlist[-1])
#print(i)
#()+1

outfile = "hives snapshot " + today +".avi"
video_index = 0
fourcc = cv2.VideoWriter_fourcc(*'XVID') # Be sure to use lower case
out = cv2.VideoWriter(outfile, fourcc, 25.0, (width, height))

for video_index in range(len(videofiles)):
    print ("processing " + str(video_index) + str(videofiles[video_index]))
    cap = cv2.VideoCapture(videofiles[video_index])
    frameCount = cap.get(cv2.CAP_PROP_FRAME_COUNT) #doesn't seem to work
    #print("frameCount is " + str(frameCount))
    firstFrame = False
    frameN = 0
    ret, frame = cap.read()
    dim = rescale(frame)
    print(dim)
    while ret:
        frame = cv2.resize(frame, dim, interpolation=cv2.INTER_AREA)
        frame = addBorder(frame,dim)
        frame = addCaption(frame,str(videofiles[video_index]))
        frame = cv2.resize(frame, (width,height), interpolation=cv2.INTER_AREA)
        cv2.imshow('frame',frame)
        out.write(frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        ret, frame = cap.read()        
        frameN += 1
    print("processed " + str(frameN))

def getDuration(f):
    vidcapture = cv2.VideoCapture(f)
    fps = vidcapture.get(cv2.CAP_PROP_FPS)
    totalNoFrames = vidcapture.get(cv2.CAP_PROP_FRAME_COUNT);
    durationInSeconds = float(totalNoFrames) / float(fps)
    return durationInSeconds
        
def junk():
        
    while True:    
        if frame is None or ret is False:
            video_index += 1
            if video_index >= len(videofiles):
                print("here")
                cap.release()
                break
            cap = cv2.VideoCapture(videofiles[ video_index ])
            frameN = 0
            ret, frame = cap.read()
        if (firstFrame == False):
            dim = rescale(frame)
            firstFrame = True
        if (video_index > 3):
            print("broke by index")
            break
        if (frameN > 39):
            frameN = 0
            print("broke by frameN")
            break
        else:
            frameN += 1

        
cap.release()
out.release()
cv2.destroyAllWindows()

print("now transcoding a file of duration " + str(getDuration(outfile)) )
transcode(outfile,outfile + ".mp4")
send2trash(curDir + outfile)
print ("end.")
```
